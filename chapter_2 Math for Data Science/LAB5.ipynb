```python
#a=customer who visited website
#b=customer who purchased
#find comman users (a n b)
#find unique users (a u b) and missing users (a-b)
#perform classificn mapping-filtering
#   ->a=predicted buyers(assuse visited are buyers)
#   ->b=actual buyers

vis_user={"alice","bob","charlie","david","eva"}
pur_user={"charlie","eva","frank","george"}
#commomn user
print(" common users: ",vis_user & pur_user)
print(" unique users: ",vis_user | pur_user)
#missing user means users who didnt purchase
print(" missing user: ",vis_user-pur_user)
#unexpected case : user who didnt visit but purchased
print(" unexpected: ",pur_user-vis_user)
#4,
pred=vis_user
act=pur_user
print(" true positive: ",pred&act)
print("false positive: ",pred-act)
print("false negetive: ",act-pred)
```

```python
#pandas implementation- ecommerce. visit data, purchase data
import pandas as pd
#pandas create dataframe
#below is visit dataset
visits_data={
    "CustomerID":["C1","C2","C3","C4","C5","C6"],
    "Visit_Count":[5,3,8,2,7,4]
}
#below is purchase dataset
purchase_data={
    "CustomerID":["C3","C4","C5","C7"],
    "Purchase_Amount":[200,150,300,250]
}

visits_df=pd.DataFrame(visits_data)
purchase_df=pd.DataFrame(purchase_data)
print("\nVisits data:\n",visits_df)
print("\Purchase data:\n",purchase_df)

visited_users=set(visits_df["CustomerID"])
purchased_users=set(purchase_df["CustomerID"])

#1. common users-AND
common=visited_users&purchased_users
print("Common users:",common)
#2. all users
allusers=visited_users | purchased_users
print("All usersP:",allusers)
#3. non buyer
nonb=visited_users-purchased_users
print("NON buyers:",nonb)
#4.buyer dint visit
buyer=purchased_users-visited_users
print("Buyer not visited:",buyer)

#ML-analysis
predicted=visited_users
actual=purchased_users
tp=predicted & actual
fp=predicted-actual
fn=actual-predicted
print("\n----Classification result----")
print("True positive: predicted the buyers:",tp)
print("False positive : predicted but dint buy:",fp)
print("False negative: dint predict but bought: ",fn)
#1.common cust-dataframes 
inner = pd.merge(visits_df, purchase_df, on="CustomerID", how="inner")
print(inner)
#visitors and pur details
left = pd.merge(visits_df, purchase_df, on="CustomerID", how="left")
print(left)
right = pd.merge(visits_df, purchase_df, on="CustomerID", how="right")
print(right)
#find non buyer
non_buyer_df=left[left["Purchase_Amount"].isna()]
print(" visitors who didnt buy: ",non_buyer_df)

#find total visited userr ,total purchased, total unique user ---> finding cardinality
print("\n")
print(len(allusers))
print("\n")
print(len(purchased_users))
```
