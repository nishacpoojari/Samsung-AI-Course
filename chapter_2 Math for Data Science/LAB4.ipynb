```python
#predict house prices
#find error prediction for each val
#cal mean absolute error and interrpt whether the model working well

import numpy as np
y_true=np.array([50,60,55,70,65])
y_pred=np.array([52,58,54,68,66])
#cal error
errors=np.abs(y_true-y_pred)
#cal mean absolute error
mae=np.mean(errors)

print(" mean error: ",mae)
```

```python
#salary data in thousands
sal=np.array([25,20,28,27,29,31,100,5])
#cal mean
mean_sal=np.mean(sal)
#cal deviation
dev=np.abs(sal-mean_sal)
#define threshold(2 times deviation))
threshold=2*np.mean(dev)
#identify outliers
out=sal[dev > threshold]

print(" sal: ",sal)
print(" mean sal: ",mean_sal)
print(" deviation: ",dev)
print("\n threshold: ",threshold)
print("out: ",out)

#modify by adding lower val like 5 , change threshold (1.5x ,3x) compare result

# different thresholds
for k in [1.5, 2, 3]:
    Threshold = k * np.mean(dev)
    out = sal[dev > threshold]
    
    print("\n--- Threshold:", k, "x ---")
    print("Threshold value:", Threshold)
    print("Outliers:", out)
```

```python
from numpy.linalg import norm
# exixting marks in 2 subjects
grp_a=np.array([[40,50],[42,48],[38,52]])
grp_b=np.array([[70,80],[77,78],[68,82]])
# new student
new_stud=np.array([45,55])
#compare dist from each group
dis_a=[norm(new_stud- point) for point in grp_a]
dis_b=[norm(new_stud- point) for point in grp_b]

print(" distance from group a: ",dis_a)
print(" distance from group b: ",dis_b)
#decision
if min(dis_a) < min(dis_b):
    print(" belongs to group a")
else:
    print(" belongs to group b")
```

```python
#data points (income,spending)
points=np.array([[2,3],[3,4],[8,9],[9,10]])
#initial centeroids
cent1=np.array([2,3])
cent2=np.array([9,10])

for point in points:
    d1=np.linalg.norm(point-cent1)
    d2=np.linalg.norm(point-cent2)

    if d1<d2:
        print(point,"->cluster 1")
    else:
        print(point,"->cluster 2")
```
