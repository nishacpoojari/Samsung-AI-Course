```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

#step 2: Rael dataset(text data)
documents = [
    "AI for agriculture using sensors",
    "Smart farming using IoT sensors",
    "Deep learning for image recognition",
    "Crop monitoring using AI and IoT"
]

#Step 3: Convert text -> numerical vectors (TF-IDF)
vectorizer = TfidfVectorizer()
tfid_matrix = vectorizer.fit_transform(documents)

#Step 4: Compute cosine similarity
cos_sim = cosine_similarity(tfid_matrix)

#Step 5: Print similarity matrix
print("Cosine similarity Matrix:\n")
print(cos_sim)

plt.imshow(cos_sim)
plt.colorbar()
plt.show()
```

```python
#ed tech and analysis- slope using tan
import numpy as np
import matplotlib.pyplot as plt
hours=np.array([2,4,6,8])
marks=np.array([45,55,65,80])
#compute slope using regression
slope,intercept=np.polyfit(hours,marks,1)
#angle
theta=np.degrees(np.arctan(slope))
print("Slope (tan theta):",slope)
print("Angle:",theta)

#plot
plt.scatter(hours,marks)
plt.plot(hours,slope*hours+intercept)
plt.title("Student performance trend")
plt.xlabel("Hours studied")
plt.ylabel("marks")
plt.show()
```
