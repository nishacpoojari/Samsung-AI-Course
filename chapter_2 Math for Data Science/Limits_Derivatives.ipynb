```python
import numpy as np
import matplotlib.pyplot as plt

def loss(w):
    return (w-3)**2

def gradient(w):
    return 2*(w-3)

#values for curve
w_values = np.linspace(-1, 6, 100)
loss_values = loss(w_values)

#start point
w=0
learning_rate = 0.2

#plot curve
plt.figure()
plt.plot(w_values, loss_values)
plt.title("Gradient Descent = Learning Direction")
plt.xlabel("Weight(w)")
plt.ylabel("Loss")
```
