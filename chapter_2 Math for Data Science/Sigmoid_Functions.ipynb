```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)

#Linear function(Before Transformation
y_linear = x  #Example: raw output

#Sigmoid transformation(After)
y_sigmoid = 1/(1+np.exp(-x))

#plot both graphs
plt.figure(figsize=(10,5))

#Before (linear)
plt.subplot(1,2,1)
plt.plot(x,y_linear)
plt.title("Before Transformation(Linear Output)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

#After (Sigmoid)
plt.subplot(1,2,2)
plt.plot(x,y_sigmoid)
plt.title("After Sigmoid Transformation")
plt.xlabel("x")
plt.ylabel("Probability")
plt.grid()

               
```

```python
import numpy as np
import matplotlib.pyplot as plt

hours = np.array([1,3.5,5,7])

#linear model (raw output)
z=1.2*hours-4

#Sigmoid function
sigmoid = 1/(1+np.exp(-z))

#Output
for i in range(len(hours)):
    if(sigmoid[i]>0.5):
        x="Pass"
    else:
        x="Fail"
    print(f"Hours:{hours[i]}, Raw : {z[i]:.2f}, Probability: {sigmoid[i]:.3f}, {x}")



```
