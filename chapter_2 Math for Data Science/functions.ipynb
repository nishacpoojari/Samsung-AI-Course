```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-5,5,1000)

y=3*x+1

plt.plot(x,y, label='f(x)=3x+1')

plt.xlabel('x (Domain)')
plt.ylabel('f(x) (Range)')
plt.title('Graph of f(x) = 3x+1')

plt.grid()
plt.legend()

plt.show()
```

```python
#composite functions

import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(-5,5,1000)

g=x**2
f_of_g = 3*(g) + 1

plt.plot(x, g, label='g(x) = x^2')
plt.plot(x, f_of_g, label='f(g(x)) = 3x^2 +1')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Composite function')
plt.grid()
plt.legend()
plt.show()
```

```python
#inverse function

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
y=3*x+1
y_inv = (x-1)/3

plt.plot(x,y, label='f(x)=3x+1')
plt.plot(x, y_inv, label='f^-1(x) = (x-1)/3')
plt.plot(x, x, linestyle = '--', label='y=x(mirror line)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and Inverse function')
plt.grid()
plt.legend()
plt.show()
```

```python
#linear function

import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4,5])
y=2*x+3

plt.scatter(x,y)
plt.plot(x,y)
plt.title("Linear function")

plt.xlabel('Input')
plt.ylabel('Output')
plt.show()



```

```python
#Linear regression example 
#year of experience and salary prediction

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x=np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([3,4,5,6,7])

model = LinearRegression()

model.fit(x,y)

predicted_salary = model.predict([[6]])

print("Predicted salary for 6 years:", predicted_salary[0])

plt.scatter(x, y, label="Actual Data")
plt.plot(x,model.predict(x),label="Regression Line")
plt.ylabel("Salary (Lakhs)")
plt.title('Linear Regression Example')
plt.legend()
```

```python
#Polyunomial function
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-5,5,100)
y = x**2 + 2*x+1
plt.plot(x,y, label='f(x)=x**2+2*x+1')
plt.title("Polynomial function")
plt.show()
```
