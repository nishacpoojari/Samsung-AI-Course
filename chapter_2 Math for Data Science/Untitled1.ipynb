```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve, latex
from IPython.display import display, Math

# Method 1: Simple string formatting
def print_linear_equation(m, b):
    """Print a linear equation in the form y = mx + b"""
    if m == 0:
        equation = f"y = {b}"
    elif m == 1:
        equation = f"y = x + {b}" if b != 0 else "y = x"
    elif m == -1:
        equation = f"y = -x + {b}" if b != 0 else "y = -x"
    else:
        equation = f"y = {m}x + {b}" if b != 0 else f"y = {m}x"
        
    # Handle negative b for cleaner display
    if b < 0:
        equation = equation.replace("+ -", "- ")
        
    print(equation)
    return equation
m = 2  # slope
b = -3  # y-intercept

# Method 1: Simple string formatting
print_linear_equation(m, b)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values
x = np.arange(0, 100)

# Define equation y = 2x + 1
y = -3 * x + -4

# Plot the graph
plt.plot(x, y, label="y = 2x + 1")

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph of Linear Equation")

# Show legend
plt.legend()

# Show grid
plt.grid()

# Display graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameter t
t = np.linspace(-10, 10, 100)

# Point r0 = (1, 2)
x0, y0 = 1, 2

# Direction vector v = (2, 1)
vx, vy = 2, 1

# Parametric equations
x = x0 + vx * t
y = y0 + vy * t

# Plot line
plt.plot(x, y, label="Vector Line")

# Plot starting point
plt.scatter(x0, y0, color='red', label="Point (2,4)")
plt.quiver(x0, y0, vx, vy, angles='xy', scale_units='xy', scale=1, color='green')
# Labels and grid
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line using Vector Form")
plt.legend()
plt.grid()

plt.show()
```
