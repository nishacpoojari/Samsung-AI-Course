```python
#Skewed data tranformation

import numpy as np
import matplotlib.pyplot as plt

data = np.array([1,10,100,1000])
log_data = np.log(data)

#data = np.array([1,5,10,50,100,500,1000])
#Reduces large variance - Better ML Performance

print("Original:",data)
print("After log:",log_data)

plt.plot(data,log_data)
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from  sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data = {
    "Size":[500,800,1000,1500,2000,3000,5000],
    "Price":[100000,150000,200000,300000,500000,900000,2000000]
}
df=pd.DataFrame(data)

plt.figure()
plt.scatter(df["Size"],df["Price"])
plt.title("Original data (Highly Skewed)")
plt.xlabel("Size")
plt.ylabel("Price")
plt.grid()
plt.show()

#apply natural log transformation        
df["Log_Price"]=np.log(df["Price"])
#visualize
plt.figure()
plt.scatter(df["Size"],df["Log_Price"])
plt.title("Log Transformed Data(Reduced Skewness)")
plt.xlabel("Size")
plt.ylabel("Log(Price)")
plt.grid()
plt.show()

X= df[["Size"]]
y= df["Price"]

model_original = LinearRegression()
model_original.fit(X,y)

pred_original = model_original.predict(X)

y_log= df["Log_Price"]

model_log = LinearRegression()
model_log.fit(X,y_log)
pred_log = model_log.predict(X)

pred_log_exp = np.exp(pred_log)


mse_original = mean_squared_error(y,pred_original)
mse_log = mean_squared_error(y,pred_log_exp)

print("MSE (Original Data):",mse_original)
print("MSE(Log Transformed Data):",mse_log)

#Plot comparision
plt.figure()
plt.scatter(X,y,label="Actual")
plt.plot(X,pred_original,label="Original Mode")
plt.plot(X, pred_log_exp, label="Log Model(Better Fit)")
plt.legend()
plt.title("Model Comparision")
plt.xlabel("Size")
plt.ylabel("Price")

```
