```python
import random
rolls = [random.randint(1, 6) for _ in range(20)]
even_numbers = {2, 4, 6}
odd_numbers = {1, 3, 5}
prime_numbers = {2, 3, 5}
even_count = sum(1 for roll in rolls if roll in even_numbers)
odd_count = sum(1 for roll in rolls if roll in odd_numbers)
prime_count = sum(1 for roll in rolls if roll in prime_numbers)
print(f"Rolls: {rolls}")
print(f"Even numbers count: {even_count},probability ={(even_count/20)}")
print(f"Odd numbers count: {odd_count},probability ={(odd_count/20)}")
print(f"Prime numbers count: {prime_count},probability ={(prime_count/20)}")
```

```python
s_set=['H','T']
coins=[random.choice(s_set) for i in range(20)]
head_count=sum(1 for toss in coins if toss=='H')
print('toss:',coins)
print("\n head count ",head_count)
```
