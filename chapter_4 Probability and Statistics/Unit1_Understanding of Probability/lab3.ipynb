```python
import numpy as np
from scipy.stats import binom, poisson

print("=" * 60)
print("AI/ML APPLICATION: SPAM EMAIL DETECTION")
print("=" * 60)

# BERNOULLI DISTRIBUTION
print("\n1. BERNOULLI DISTRIBUTION")
print("-" * 40)

# Probability that an email is spam
p_spam = 0.2

# Generate one email
email = np.random.binomial(1, p_spam)

print("Probability of spam =", p_spam)

if email == 1:
    print("Result: The email is SPAM")
else:
    print("Result: The email is NOT SPAM")

print("\nInterpretation:")
print("Bernoulli Distribution models a SINGLE email.")
print("Possible outcomes:")
print("1 -> Spam")
print("0 -> Not Spam")

# BINOMIAL DISTRIBUTION
print("\n\n2. BINOMIAL DISTRIBUTION")
print("-" * 40)

n = 15  # Total emails
x = 4   # Required spam emails

binomial_probability = binom.pmf(x, n, p_spam)

print("Total Emails (n) =", n)
print("Probability of Spam (p) =", p_spam)
print("Required Spam Emails (x) =", x)

print("\nProbability of exactly 4 spam emails:")
print("P(X=4) =", round(binomial_probability, 6))

print("\nInterpretation:")
print("Binomial Distribution models MULTIPLE emails.")
print("It calculates the probability of getting")
print("exactly 4 spam emails out of 15 emails.")

# POISSON DISTRIBUTION
print("\n\n3. POISSON DISTRIBUTION")
print("-" * 40)

lam = 3  # Average spam emails per minute
x = 5    # Required spam emails

poisson_probability = poisson.pmf(x, lam)

print("Average Spam Emails per Minute (λ) =", lam)
print("Required Spam Emails (x) =", x)

print("\nProbability of receiving exactly 5 spam emails:")
print("P(X=5) =", round(poisson_probability, 6))

print("\nInterpretation:")
print("Poisson Distribution models the number of spam")
print("emails arriving in a fixed time interval.")
print("Here, it calculates the probability of receiving")
print("exactly 5 spam emails in one minute when the")
print("average arrival rate is 3 spam emails per minute.")

print("\n" + "=" * 60)
print("SUMMARY")
print("=" * 60)
print("Bernoulli -> Single email (Spam / Not Spam)")
print("Binomial  -> Number of spam emails in a batch")
print("Poisson   -> Spam email arrivals over time")
print("=" * 60)
```
