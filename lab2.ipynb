```python
# Loan approval prediction model (Naive Bayes approach)
# Prior probabilities
p_repay = 0.70
p_default = 0.30

# Likelihoods: P(Feature | Class)
# Example feature: High Credit Score (HCS)
p_hcs_given_repay = 0.85    # P(High Credit Score | Repay)
p_hcs_given_default = 0.20  # P(High Credit Score | Default)

# Evidence: P(High Credit Score)
# Total Probability Theorem: P(A) = P(A|B1)*P(B1) + P(A|B2)*P(B2)
p_high_credit_score = (p_hcs_given_repay * p_repay) + (p_hcs_given_default * p_default)

# Posterior probabilities: P(Class | Feature)
# Bayes' Theorem: P(Class|Feature) = (P(Feature|Class) * P(Class)) / P(Feature)
p_repay_given_hcs = (p_hcs_given_repay * p_repay) / p_high_credit_score
p_default_given_hcs = (p_hcs_given_default * p_default) / p_high_credit_score

# Output results
print(f"Total Probability of High Credit Score: {p_high_credit_score:.4f}")
print(f"Probability of Repayment given High Credit Score: {p_repay_given_hcs:.4f}")
print(f"Probability of Default given High Credit Score: {p_default_given_hcs:.4f}")

# Decision rule
if p_repay_given_hcs > p_default_given_hcs:
    print("Decision: Approve Loan")
else:
    print("Decision: Reject Loan")
```

```python
# Prior probabilities
p_repay = 0.70
p_default = 0.30

# Likelihoods
p_hcs_given_repay = 0.85
p_hcs_given_default = 0.20

n = int(input("Enter number of clients: "))

for i in range(n):
    print("\nClient", i + 1)

    credit_score = input("High Credit Score? (yes/no): ")

    if credit_score.lower() == "yes":
        p_feature = (p_hcs_given_repay * p_repay) + (p_hcs_given_default * p_default)

        p_repay_given = (p_hcs_given_repay * p_repay) / p_feature
        p_default_given = (p_hcs_given_default * p_default) / p_feature
    else:
        p_lcs_given_repay = 1 - p_hcs_given_repay
        p_lcs_given_default = 1 - p_hcs_given_default

        p_feature = (p_lcs_given_repay * p_repay) + (p_lcs_given_default * p_default)

        p_repay_given = (p_lcs_given_repay * p_repay) / p_feature
        p_default_given = (p_lcs_given_default * p_default) / p_feature

    print(f"Repayment Probability: {p_repay_given:.4f}")
    print(f"Default Probability: {p_default_given:.4f}")

    if p_repay_given > p_default_given:
        print("Decision: Approve Loan")
    else:
        print("Decision: Reject Loan")
```

```python
import random

p=0.3

n=20

emails=[]

for i in range(n):
    email = 1 if random.random() < p else 0
    emails.append(email)

print("email classification results: ")
print("Generated Data:",emails)

mean = sum(emails)/len(emails)
variance = sum((x-mean)**2 for x in emails)/len(emails)
print("estimated probability of spam(mean):", round(mean,3))
print("variance:",round(variance,3))

print("\nTotal Spam Emails:", sum(emails))
print("Total Non-Spam emails:",n-sum(emails))
```
