```python
#Checking and processing duplicate data

import pandas as pd
df = pd.DataFrame({'c1':['a','a','b','a','b'],
                   'c2':[1,1,1,2,2],
                   'c3':[1,1,2,2,2]})
print(df)

df_dup = df.duplicated()    #finding if any row is duplicated
print(df_dup)

col_dup = df['c2'].duplicated()    #finding duplicate values in specific columns
print(col_dup)

df2 = df.drop_duplicates()
print(df2)

df3 = df.drop_duplicates(subset=['c2','c3'])
print(df3)
```
