```python
dat = sns.load_dataset('iris')

sns.pairplot(dat)
plt.show()
```

```python
sns.pairplot(dat, hue='species')
plt.plot()
```

```python
sns.pairplot(dat,hue='species',palette='coolwarm')
plt.show()
```

```python
g= sns.PairGrid(dat)
g.map(plt.scatter)
plt.show()
```

```python
g=sns.PairGrid(dat)
g.map_diag(sns.distplot, color='orange')
g.map_upper(plt.scatter, color = 'green')
g.map_lower(sns.kdeplot, color='blue')
plt.show()
```

```python
g = sns.FacetGrid(data=dat, col='species')
g.map(sns.distplot, 'petal_length')
plt.show()
```

```python
g = sns.FacetGrid(data=dat, row='species')
g.map(sns.distplot, 'petal_length')
plt.show()
```

```python
g = sns.FacetGrid(data=dat, col='species')
g.map(plt.scatter,'petal_length', 'petal_width',color='orange')
plt.show()
```
