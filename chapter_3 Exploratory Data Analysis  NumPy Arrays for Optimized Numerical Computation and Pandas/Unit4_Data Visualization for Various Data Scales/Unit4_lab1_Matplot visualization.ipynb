```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.array(['Q1','Q2','Q3','Q4'])
y = np.array([234.0,254.7, 144.6, 317.6])
plt.bar(x,y,color = 'purple')
plt.title('BAR CHART')
plt.show()
```

```python
x = np.random.randn(10000)
plt.hist(x,bins=20, color='green',density = True)
plt.title('HISTOGRAM')
plt.show()
```

```python
x = np.random.randn(10000)
plt.hist(x, bins=100, color='blue', density = False, alpha=0.3)
plt.title('HISTOGRAM')
plt.show()
```

```python
x = np.random.randn(10000)*10+3
y = np.random.randn(10000)*10+5
z = np.random.randn(10000)*10+1
plt.boxplot([x,y,z],0)
plt.title('BOX PLOTS')
plt.show()

plt.boxplot([x,y,z],0,vert = False)
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
#plt.plot(x,y,color='orange', linewidth=2)
#plt.plot(x,y,color='green',linestyle=':',linewidth=2)
#plt.plot(x,y,color='green',linestyle='-.',linewidth=3)
plt.plot(x,y,color='green',linestyle='--',linewidth=2)

plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.show()
#help(plt.plot)
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='green',linestyle='-',linewidth=2)

plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')
plt.xlim([0,5])
plt.ylim([-2,+2])
plt.show()

```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

fig0 = plt.figure()
axes0 = fig0.add_axes([0,0,1,1])     #left, bottom, width, height
#axes0.plot(x,y,color='red', linestyle='-')
axes0.plot(x,y,color='purple', linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
axes0.set_xlim([0,5])
axes0.set_ylim([-2,2])

plt.show()
```

```python
fig0 = plt.figure(figsize=(10,1))
axes0 = fig0.add_axes([0,0,1,1]) 
axes0.plot(x,y,color='red', linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
plt.show()
```

```python
fig0 = plt.figure(figsize=(5,2),dpi=100)
axes0 = fig0.add_axes([0,0,1,1]) 
axes0.plot(x,y,color='blue', linestyle='--',label='Sin')
axes0.plot(x,z,color='purple', linestyle='-.',label='Cos')
axes0.legend(loc=0)
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
plt.show()
```

```python
fig0 = plt.figure(figsize=([8,3]),dpi=80)   #Width, height, and DPI setting
axes1 = fig0.add_axes([0,0,0.3,1]) 
axes2 = fig0.add_axes([0.5,0,0.3,1]) 

#axes1.plot(x,y,color='blue', linestyle=':')
axes1.plot(x,y, marker ='^', linestyle='none',markersize=10, markerfacecolor = 'purple', markeredgecolor = 'red', alpha =0.5)
axes2.plot(x,z,color='purple', linestyle='-.')

axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('SINE')

axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('COSINE')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
# (0,0)
c= np.random.randn(10000)
axes[0,0].hist(c,bins=100,color='green',density=True,alpha=0.3)
axes[0,0].set_title('HISTOGRAM')
# (1,0)
axes[1,0].plot(x,z, linestyle='none',marker='^',markersize=10,markerfacecolor='purple',markeredgecolor='red',alpha=0.5)
axes[1,0].set_xlabel('X')
axes[1,0].set_ylabel('Y')
axes[1,0].set_title('COSINE')
# (0,1)
a = np.array(['01', '02', '03','04'])
b = np.array([ 234.0, 254.7, 144.6, 317.6])
axes[0,1].bar(a,b,color='blue',alpha=0.5)
axes[0,1].set_title('BAR PLOT')
# (1,1)
axes[1,1].plot(x,y,color='green', linestyle=':')
axes[1,1].set_xlabel('X')
axes[1,1].set_ylabel('Y')
axes[1,1].set_title('SINE')

plt.tight_layout()
plt.show()

# Avoid overlapping.








```
