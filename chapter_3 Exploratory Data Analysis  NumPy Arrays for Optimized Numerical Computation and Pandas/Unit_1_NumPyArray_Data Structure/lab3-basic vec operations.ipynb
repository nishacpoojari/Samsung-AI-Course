```python
x=np.arange(1,11)
print(x)
print(x.sum())
print(x.cumsum())
print(x.mean())
print(x.std())
print(x.var())
print(x.max())
print(x.min())
#find loc of max and min
print(x.argmax())
print(x.argmin())
```

```python
x=np.arange(1,11)
x_mean=x.mean()
print(x_mean)
dev=x-x_mean
print(dev)
sqdev=dev**2
print(sqdev)
sumsqdev=sqdev.sum()
print(sumsqdev)
var=sumsqdev/x.size
print(var)
print(x.var())
sd=np.sqrt(var)
print(sd)
print(x.std())
```

```python
np.random.seed(123)
a=np.random.randint(1,11,size=1000)
print(a)
print(a.size)
print(np.unique(a))
print(np.median(a))
print(np.sum(a))
print(np.round(np.var(a),2))
print(np.round(np.mean(a),2))
print(np.round(np.std(a),2))
print(np.max(a))
print(np.min(a))
#first occurance of max and min
print(np.argmax(a))
print(np.argmin(a))
```
